### Concat two DataFrames

```python
df1=pd.DataFrame({
    'Name':["Akash",'Priysnka','Rahul'],
'Marks':[4,5,6]})

df2=pd.DataFrame({
    'Name':["Akash",'Priysnka','Rahul'],
'Roll No':[4234,2345,654]})

pd.concat([df1,df2])
```

### Merging two df
- `on='Col-Name'` 
    - merging two df on Col-Name column 
    - return only duplicate rows 


```python
pd.merge(df1,df2,on='Name')
```
