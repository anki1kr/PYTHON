# Analysis of zomato sales

```python
%pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [, ]
sales_data = np.array([
    [0,2021, 2022, 2023, 2024],           #year
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
                                 
print("\n Sample data for top 3 restau:\n", sales_data[1:4,1:])
                #[slice rows1,2,3 only,|1 exception:all cols]
```

### Min,Max,Avg

```python
print("NP sum",sum:=np.sum(sales_data[1:,1:],axis=0))

print("Yearly total :",yearlyTotal:=np.sum(sales_data[1:,1:],axis=0))                             

print("MINIMUM Sales per restu:",minSales:=np.min(sales_data[1:,1:],axis=1))

print("MAXIMUM Sales per yr:",maxSales:=np.max(sales_data[:,1:],axis=0))

print("AVG Sales per restu:",avgSales:=np.mean(sales_data[:,1:],axis=1))
```

### Plotting Cumsum

```python

print("Cumulative sum\n",cumsum:=np.cumsum(sales_data[1:,1:],axis=1))

print("AvgCumSales: ",avgCumSales:=np.mean(cumsum,axis=0))

plt.figure(figsize=(7,6))
plt.plot(avgCumSales)
plt.title("Average Cumultive Sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Properties

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

## Handling missing values

```python
"""Return True if x is a NaN (not-a-number) value.
and we can't compare is_nan

nan_to_num converts nan value to number (by default 0)

isinf() (is infinity) true or false"""
arr = np.array([[1,np.nan, 3],
                [4, np.nan, 6]])

print("is_nan\n",np.isnan(arr))

print("\nnan_to_num\n",np.nan_to_num(arr))
print("nan_to_num\n",np.nan_to_num(arr,nan=100))
```

### IS infinity and handle it

```python
inf_arr = np.array([[1, np.inf, 3],
                    [4, -np.inf, 6]])
print("\nisinf\n", np.isinf(inf_arr))

print("\nReplace inf with \n", np.nan_to_num(inf_arr,posinf=1000,neginf=-1000))
```
