### Array properties

```python
arr=np.array([[1.5,2.7,3.4],
              [4,5.2,6]])
print("Dimensions:",arr.ndim)
print("Size:",arr.size) #totalm num of elements
print("Shape:",arr.shape)
print("data-type:",arr.dtype)
print("Conversion into float:\n",arr:=arr.astype(int))
```

### Array operation

```python
print("\nadditon\n",arr+2)
print("\nSub\n",arr-1)
print("\nMul\n",arr*2)
print("\nPow\n",arr**2)
```

### Aggregate Functions

```python
arr=np.arange(10,60,10)#(start,stop,step)
print("\nOrigional array:\n",arr)
print(np.sum(arr))
print("Variation :",np.var(arr))
print(np.min(arr))
print(np.max(arr))
print("Standard deviation :",np.std(arr))

```

### Array reshaping

```python
arr=np.arange(16)
print("\nOrigional array:\n",arr)

reshaped=arr.reshape((4,4))
print("\nReshaped array:\n",reshaped)

flatten=arr.flatten()
print("\nFlatten array:\n",flatten)

#return view, instead of copy
raveled=arr.ravel()
print("\nravled array:\n",raveled)

transpose=reshaped.T
print("\nTransposed array:\n",transpose)
```

### Slicing

```python
arr=np.arange(12)
print("Print Array:",arr)
print("\n Basic Slicing:",arr[2:9])
print("\n Step Slicing:",arr[1:10:2])
print("\n Reversed using Negetaive idx Slicing:",arr[::-1]) #(from start::last)
print("\n Negetaive idx Slicing:",arr[-1])
```

### Sorting


```python
unsorted=np.array([5,1,6,8,2,9,3])
print("Sorted 1D Array:",np.sort(unsorted))

arr2D=np.array([[5,3],
                [2,1],
                [3,1]])
print("\nSorted 2D array:\n",np.sort(arr2D,axis=0)) #column wise sorted

print("\nSorted 2D array:\n",np.sort(arr2D,axis=1)) #row wise sorted
```

### Adding rows or column

```python
original_arr=np.array([[1,2],
                       [5,6],
                       [9,10]])
new_row=np.array([12,13])
        #Vertical stack < axis0
new_col=np.array([[3],
                  [7],
                  [11],
                  [14]])

print("\nwith new row added\n",withNewRow:=np.vstack((original_arr,new_row)))
            #horizontal stack < axis 1

print("\nwith new col added\n",withNewCol:=np.hstack((withNewRow,new_col)))
```

### Deletion

```python
deleted=np.delete(withNewRow,0) #removes after fallterning the array
print("After deletion\n",deleted)

arr=np.array([[1, 2, 3],
			  [5, 6, 7],
			  [9, 10, 11],
			  [12, 13, 14]])
print("Deletion in 2D:\n",np.delete(arr,1,axis=0))
print("Deletion in 2D with axis 1:\n",np.delete(arr,2,axis=1))
```
