### Numpy array vs list


```python
import time

import numpy as np

start =time.time()
py_list = [i*2 for i in range(100000)]
print("\nPython list execution time: ",time.time()-start)

start=time.time()
np_array=np.arange(10000)*2
print("Numpy array operation time: ",time.time()-start)
```

### Matrix

```python
zero=np.zeros((3,3))
print("\nZero array: \n",zero)

ones=np.ones((3,3))
print("\nOnes array: \n",ones)

full=np.full((3,3),5)
print("\nConstant element in array: \n",full)

random=np.random.random((3,3))
print("\nRandom elements in  array: \n",random)

sequence=np.arange(1,12,2)
print("\nSequenced Array: \n",sequence)
```

### Vector ,Matrix and tensor


```python
vector=np.array([1,2,3,4])
print("Vector: \n",vector)

matrix=np.array([[1,2,3],
                [4,5,6]])
print("\nMatrix: \n",matrix)

tensor=np.array([ [[1,2],[3,4]],

                [[4,5],[6,7]]    ])
print("\nTensor: \n",tensor)
```

### Array properties

```python
arr=np.array([[1,2,3],
              [4,5,6]])
print("Dimensions:",arr.ndim)
print("Size:",arr.size)
print("Shape:",arr.shape)
print("data-type:",arr.dtype)
```

### Array reshaping

```python
arr=np.arange(16)
print("\nOrigional array:\n",arr)

reshaped=arr.reshape((4,4))
print("\nReshaped array:\n",reshaped)

flatten=arr.flatten()
print("\nFlatten array:\n",flatten)

#return view, instead of copy
raveled=arr.ravel()
print("\nravled array:\n",raveled)

transpose=reshaped.T
print("\nTransposed array:\n",transpose)
```


##Phase 2
### Slicing


```python
arr=np.arange(12)
print("Print Array:",arr)
print("\n Basic Slicing:",arr[2:9])
print("\n Step Slicing:",arr[1:10:2])
print("\n Reversed using Negetaive idx Slicing:",arr[::-1]) #(from start::last)
print("\n Negetaive idx Slicing:",arr[-1])
```

### 2D matrices
